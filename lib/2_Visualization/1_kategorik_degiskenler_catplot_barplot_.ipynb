Hemen aşağısında çok daha açıklayıcı olabilecek catplot'u kullanacağız.

```python
sns.catplot(x = "writer", y = "reading_time", data = df);
```

### Görüldüğü üzere mesela barplotta şu bilgi çıkmıyor; dostoyevskiyi en az 60dk okumuşum, platonu sadece yarım saatler şeklinde okumuşum, barplotta dümdüz kutu var ve ortalaması var.

```python
# Kitaplara göre okuma sürem

sns.catplot(x = "book", y = "read_pages", data = df);
```

```python
# Hangi yazardan hangi kitabı kaçar sayfa okumuşum

sns.catplot(x = "writer", y = "read_pages", hue= "book" ,data = df);
```

Eğer verisetimiz bu kadar küçük olmasa ve aynı yazardan farklı kitaplar okumuş olsaydım bu grafik harika gözükecek, tek grafikte oldukça fazla bilgi verecekti.

```python
sns.catplot(x="writer", y="reading_time", hue="book", kind = "point", data = df);
```

Aynı şekilde verisetimiz büyük olmadığı için harcanan ama çok değerli bir grafik.
