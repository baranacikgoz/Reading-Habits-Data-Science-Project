```python
sns.FacetGrid(df,hue = "book",height = 5,xlim = (0,75)).map(sns.kdeplot, "read_pages", shade = True).add_legend();

# Koddaki bileşenlerin ne işe yaradığı;
# FacetGrid=grafik uzerine eklenen boyutlari bolerek gostermek icin
# xlim= 0 koyarak, grafiğin eksi kısmını almadık, 75 de mesela max okduğum değere falan denk geliyor fazlasına gerek yok
# ---
# Üstteki kod sunu ifade ediyor;
# read_pages'e gore bir kdeplot grafigi olusturagiz,
# FacetGrid kullanarak tum "book" degiskeninin siniflariyla boyutlandirmis olacagiz.
# bunu yapmanin yolu da map fonksiyonu
# add_legend = book degiskeninin siniflarinin(Utopya,Suc ve Ceza vs.) yanda isminin gözükmesi
```
