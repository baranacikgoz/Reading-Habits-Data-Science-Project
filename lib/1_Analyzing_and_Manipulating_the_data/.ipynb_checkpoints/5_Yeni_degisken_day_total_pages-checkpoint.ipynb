```python
df2 = pd.DataFrame({'date': df["date"],
                    'read_pages': df["read_pages"]})

df2
```

---
Şimdi aşağıdaki kod ile aynı gün olan değerleri toplayalım.

```python
df3 = df2.groupby("date").sum()
df3
```

---
Şimdi ikisini "date"e bağlı olarak birleştirelim.

```python
df4 = pd.merge(df2,df3, on="date")
df4
```
